```python
from selenium import webdriver
import pandas as pd 

driver = webdriver.Chrome('C:/Users/ASUS/Desktop/SocialNetworkWorkshop/chromedriver_win32/chromedriver')  # Optional argument, if not specified will search path.
driver.get('http://www.facebook.com/');
```

```python
import time
from bs4 import BeautifulSoup
import json

friends = driver.find_elements_by_class_name('uiProfileBlockContent')
last_i = 0
count_nothing = 0
while True:
    if last_i == len(friends):
        count_nothing += 1
    if count_nothing >= 3:
        print('=== End ===')
        break
    for i in range(last_i,len(friends)):
        out = friends[i].text.split('\n')[0]
        soup = BeautifulSoup(friends[i].get_attribute('innerHTML'), 'html.parser')
        id=json.loads(soup.find('a').get('data-gt'))['engagement']['eng_tid']
        url=soup.find('a').get('href')[:soup.find('a').get('href').find('?')]
        if 'profile.php' in url:
            url = url+'?id='+id
        print(out+','+id+','+url)
    last_i=len(friends)
    # for i in range(10):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    friends = driver.find_elements_by_class_name('uiProfileBlockContent')        
```
